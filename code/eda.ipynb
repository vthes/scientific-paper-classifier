```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
data_path = '../data/final_data.csv'
data = pd.read_csv(data_path)
term_counts = data.iloc[ : , 2 : ].sum(axis = 0).reset_index()
term_counts.columns = ['term', 'count']
```

```python
fig = px.histogram(term_counts[term_counts['count'] > 100], x="term", y="count", log_y=True)
fig.show()

# Only 38 terms with more than a 100 occurrences
```

```python
top_40_terms = term_counts[term_counts['count'] > 100].sort_values(by=['count'], ascending=False)[:40]
top_40_terms.reset_index(inplace=True, drop=True)
```

```python
corr_data = data[top_40_terms['term'].to_list()].corr()

fig = px.imshow(corr_data,
                x=corr_data.columns,
                y=corr_data.index
               )
fig.show()

# Some of the terms have strong positive or negative correlation
```

```python
word_counts = pd.Series(dtype='float64')

for column in data.columns[2 : ]:
    word_counts[column] = data[data[column] == 1]['abstract'].apply(len).mean()

word_counts_100 = word_counts.sort_values(ascending=False)[:100]
```

```python
fig = px.histogram(word_counts_100, x=word_counts_100.index, y=word_counts_100, labels={'index':'term', 'y':'word count'})
fig.show()
```
