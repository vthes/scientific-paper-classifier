```python
import pandas as pd
```

```python
data1_path = '../data/arxiv_data_1.csv'
data2_path = '../data/arxiv_data_2.csv'
```

```python
data1_df = pd.read_csv(data1_path)
data2_df = pd.read_csv(data2_path)
```

```python
data1_df = data1_df.rename(columns={'titles':'title','summaries':'abstract'})
data1_df.head()
```

```python
data2_df = data2_df.rename(columns={'titles':'title','abstracts':'abstract'})
data2_df = data2_df[['title', 'abstract', 'terms']]
data2_df.head()
```

```python
merged_data = pd.concat([data1_df, data2_df])
clean_data = merged_data.drop_duplicates()
clean_data.to_csv('../data/clean_data.csv')
clean_data.reset_index(inplace=True, drop=True)
clean_data
```

```python
from scraping import Scraper
scraper = Scraper('https://arxiv.org/category_taxonomy')
scraper.fetch_data()
valid_terms = scraper.get_ids()
valid_terms_df = pd.DataFrame(columns = valid_terms)
final_data = clean_data
final_data = final_data.join(valid_terms_df)
```

```python
final_data.fillna(0, inplace=True)

for column in final_data.columns[3:]:
    final_data[column].loc[final_data['terms'].apply(lambda x: True if column in x else False)] = 1
```

```python
final_data.drop('terms', inplace=True, axis=1)
```

```python
final_data.to_csv('../data/final_data.csv', index=False)
```
